```python
import numpy as np
import pandas as pd
import seaborn as sb 
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
iris=pd.read_csv("/content/Iris.csv")
iris.sample(5)
```

```python
iris.head(5)
```

```python
iris.describe()
```

```python
df_norm = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
df_norm.sample(n=5)
```

```python
df_norm.describe()
```

```python
iris["Species"].sample(n=5)
```

```python
target = iris[['Species']].replace(['Iris-setosa','Iris-versicolor','Iris-virginica'],[0,1,2])
target.sample(n=6)
```

```python
df = pd.concat([df_norm, target], axis=1)
df.sample(n=5)
```

```python
train, test = train_test_split(df, test_size = 0.3)
trainX = train[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
trainY=train.Species
testX= test[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']] 
testY =test.Species
trainX.head(5)
```

```python
clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(3, 3), random_state=1)
clf.fit(trainX, trainY)
```

```python
prediction = clf.predict(testX)
print(prediction)
```

```python
print(testY.values)
```

```python
print('Accuracy:',metrics.accuracy_score(prediction,testY)*100,"%")
```
