```python
from sklearn.datasets import load_iris
from sklearn import metrics
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
iris=load_iris()
x=iris.data
y=iris.target
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
dt=DecisionTreeClassifier()
dt=dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
print("Accuracy of test: ",metrics.accuracy_score(y_test,y_pred))
```

```python
import graphviz
from sklearn.tree import export_graphviz
from IPython.display import display
dtree=graphviz.Source(export_graphviz(dt,class_names=iris.target_names,feature_names=iris.feature_names))
display(dtree)
#plt.figure(figsize=(20,15))
#tree.plot_tree(dt,fontsize=15,class_names=iris.target_names,feature_names=iris.feature_names)
#plt.show()
```
